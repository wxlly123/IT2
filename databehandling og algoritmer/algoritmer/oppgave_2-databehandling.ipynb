```python
import json
with open("imdb_top250.json") as fil:
    data = json.load(fil)
data
```

```python
#oppgave2.1
navn = data[0]['navn']
regissører = data[0]['regissører'][0]   
print('første filmen i lista heter:', navn, 'laget av', regissører)
```

```python
#oppgave2.2
total = 0
antall = 0
for film in data:
    antall += 1
    total += film['karakter']
print('gjennomsnittskarakteren er:',round(total / antall,2))
```

```python
#oppgave2.3
total = 0
antall = 10

for n in range(0,10):
    total += data[n]['karakter']
print('gjennomsnittskarakteren er til de 10 første filmene er:',round(total / antall,2))
```

```python
#oppgave2.4
regissører = {}
for film in data:
    for regissør in film["regissører"]:
        if regissør in regissører.keys():
            regissører[regissør] += 1
        else:
            regissører[regissør] = 1
 
personer_sortert = sorted(regissører.items(), key=lambda regissør:regissør[1],reverse=True)
print(personer_sortert[:3])
```
